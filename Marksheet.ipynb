```python
import pandas as pd
```

```python
math = pd.read_csv('Maths.csv')
english = pd.read_csv('English.csv')
nepali = pd.read_csv('Nepali.csv')
science = pd.read_csv('Science.csv')
student = pd.read_csv('student.csv')
```

```python
marksheet = pd.concat([student,english, math, nepali, science], axis=1, join="inner")
```

```python
marksheet.head()
```

```python
marksheet.columns = ['Roll No', 'Name', 'Roll No', 'English', 'Roll No','Math', 'Roll No','Nepali', 'Roll No','Science']
```

```python
marksheet.head()
```

```python
marksheet=marksheet.T.drop_duplicates().T
```

```python
marksheet.head()
```

```python
marksheet['Total'] = marksheet['English'] + marksheet['Math'] + marksheet['Nepali'] + marksheet['Science']
```

```python
marksheet.head()
```

```python
marksheet['Percentage'] = (marksheet['Total'] / 400)* 100
```
